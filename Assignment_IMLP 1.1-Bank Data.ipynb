```python
 def entropy(S,target_attribute):

  d=dict(S[target_attribute].value_counts())
  result=0
  for i in d.values():
    result-= ((i)/(len(S[target_attribute])))*log2((i)/(len(S[target_attribute])))
  return result

```

```python
entropy(data,'y')
```

```python
def Information_Gain(S,attribute):

  total_entropy = entropy(S,'y')  
     
  values,counts= np.unique(data[attribute],return_counts=True)  
       
  Weighted_Entropy = np.sum([(counts[i]/np.sum(counts))*entropy(data.where(data[attribute]==values[i]).dropna(),'y') for i in range(len(values))])  
      
  Information_Gain = total_entropy - Weighted_Entropy  
  
  return Information_Gain
  
```

```python
def ID3(data,origin_data,features,target_attribute_name='y',parent_node_class = None):
  if len(np.unique(data[target_attribute_name]))<=1:
    return np.unique(data[target_attribute_name])[0]

  elif len(data)==0:
    return np.unique(origin_data[target_attribute_name])[np.argmax(np.unique(origin_data[target_attribute_name],return_counts=True)[1])]

  elif len(features)==0:  #Here we can adjust the depth of the tree by tampering with the len(feature)==X.
    return parent_node_class  
      
    #If none of the above holds true, grow the tree!  
      
  else:
    parent_node_class = np.unique(data[target_attribute_name])[np.argmax(np.unique(data[target_attribute_name],return_counts=True)[1])]  
          
    #Select the feature which best splits the dataset via List Comprehension
    info_values = [Information_Gain(data,feature) for feature in features] 
    best_feature_index = np.argmax(info_values)  
    best_feature = features[best_feature_index]  
          
    #Create the tree structure.
    tree = {best_feature:{}}  
          
          
    #Remove the feature with the best inforamtion gain from the feature space  
    features = [i for i in features if i != best_feature]  
          
  
          
    for value in np.unique(data[best_feature]):  
      #Split the dataset along the value of the feature with the largest information gain 
      sub_data = data.where(data[best_feature] == value).dropna()  
              
      #Call the ID3 algorithm for each of those sub_datasets with recursion.
      subtree = ID3(sub_data,origin_data,features,target_attribute_name,parent_node_class)  
               
      tree[best_feature][value] = subtree
    return(tree)      
```
